```python
from google.colab import files
uploaded = files.upload()
```

```python
import openai
import pandas as pd
import io


api_key = 'enter api key here'

#Load excel
df = pd.read_excel(io.BytesIO(uploaded['Dataset_Q_Human_ChatGPT.xlsx']), sheet_name= 'Dataset_Q_Human_ChatGPT')

responses = [] #response data

openai.api_key = api_key

#Go through each question in the question column and generate a response
for question in df['Question']:
    response = openai.completions.create(
        model="text-davinci-002",
        prompt=question,
        max_tokens=150,
    )
    responses.append(response.choices[0].text)

df['Responses'] = responses #create response column w/ response data

df.to_excel('Dataset_Q_Human_ChatGPT.xlsx', index=False) #save data back into excel
```
