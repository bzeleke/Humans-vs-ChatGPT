```python
import nltk
from nltk.corpus import stopwords
import re

sws = list(stopwords.words('english'))
sws.append('selftext')

for i in range(len(questions)):
    questions[i] = re.sub(r'\W+', ' ', questions[i].lower())
    questions[i] = ' '.join([word for word in questions[i].split() if word not in sws])
    
    
print(len(questions))
```

### n-grams from the questions 

```python
from nltk import ngrams
unigrams = {}
bigrams = {}
trigrams = {}

for i in range(len(questions)):
    words = questions[i].split(' ')
    unis = ngrams(words,1)
    bis = ngrams(words,2)
    tris = ngrams(words, 3)
    for un in unis:
        try:
            unigrams[un[0]]+=1
        except:
            unigrams[un[0]]=1
    for b in bis:
        try:
            bigrams[b[0]+'-'+b[1]]+=1
        except:
            bigrams[b[0]+'-'+b[1]]=1
    for t in tris:
        try:
            trigrams[t[0]+'-'+t[1]+'-'+t[2]]+=1
        except:
            trigrams[t[0]+'-'+t[1]+'-'+t[2]]=1
    

import operator
sort_uni = sorted(unigrams.items(), key=operator.itemgetter(1), reverse=True)
sort_bi = sorted(bigrams.items(), key=operator.itemgetter(1), reverse=True)
sort_tri = sorted(trigrams.items(), key=operator.itemgetter(1), reverse=True)

for i in range(10):
    print(sort_uni[i])
print('---------------------------------')
    
for i in range(10):
    print(sort_bi[i])
print('---------------------------------')
    
for i in range(10):
    print(sort_tri[i])
```
