```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('distilbert-base-nli-mean-tokens')
```

```python
#from sentence_transformers import SentenceTransformer
#model = SentenceTransformer('distilbert-base-nli-mean-tokens')
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.', 
    'The quick brown fox jumps over the lazy dog.']
sentence_embeddings = model.encode(sentences)
embeds=[]
for sentence, embedding in zip(sentences, sentence_embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    embeds.append(embedding)
    print("")

print(len(embeds))
```

```python
import pandas as pd

df = pd.read_csv('Clean_100.csv')
df.head()
```

```python
human = df['Human'].tolist()
chatgpt = df['ChatGPT'].tolist()

human_embeddings = model.encode(human)
human_embeds=[]
for sentence, embedding in zip(human, human_embeddings):
    human_embeds.append(embedding)
    print("")

print(len(human_embeds))


chatgpt_embeddings = model.encode(chatgpt)
chatgpt_embeds=[]
for sentence, embedding in zip(chatgpt, chatgpt_embeddings):
    chatgpt_embeds.append(embedding)
    print("")

print(len(chatgpt_embeds))
```

### Cosine Similarity between the embeddings 

```python
from sklearn.metrics.pairwise import euclidean_distances

matrix = euclidean_distances(human_embeds, chatgpt_embeds)

print(matrix.mean())
print(matrix.std())
```

```python
len(matrix)
```

```python
mats = []
for i in range(100):
    mats.append(matrix[i][i])
    print(matrix[i][i])
    
print(mats)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

vals = cosine_similarity(human_embeds, chatgpt_embeds)
print(vals)

print(vals.mean())
print(vals.std())
```

```python
mats2 = []
for i in range(100):
    mats2.append(vals[i][i])
    print(vals[i][i])
    
print(mats2)
print(len(mats2))
```

```python
import numpy as np

arr = np.array([0,0.25,0.5,0.75,1.1])
newvals = np.digitize(mats2,arr)
print(len(newvals))
print(newvals)
```
